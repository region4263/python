```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplots()

x = np.linspace(-100, 100, 200)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30
plt.plot(x, y , label='')

plt.xlabel(' X')
plt.ylabel(' Y')
plt.title(' ')
plt.legend()

plt.show()
```

```python
"""  ,  : f > 0, f < 0 """

x = np.linspace(-100, 100, 20000)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30

y_masked = np.ma.masked_where(y < 0, y)
plt.plot(x, y_masked, label=' > 0')
y_masked = np.ma.masked_where(y > 0, y)
plt.plot(x, y_masked, label=' < 0')

plt.legend()
plt.show()
```

```python
"""   """

x = np.linspace(0, 100, 10000)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30
plt.plot(x, y, label='  0  100')

max_y = max(y)
for x, y in zip(x, y):
    if y == max_y:
        print(f'   [0: 100]  : x={x}, y={y}')
        plt.scatter(x, y, color='r', label='')

plt.legend()
plt.show()
```

```python
"""  ,     """

x = np.linspace(50, 100, 1000)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30

ls_x = []
ls_y = []
function_up = {}
flag = 0
for x, y in zip(x, y):
    ls_x.append(x)
    ls_y.append(y)
    if len(ls_y) < 3:
        continue
    #   
    elif ls_y[-3] > ls_y[-2] and ls_y[-2] < y:
        x_min = ls_x[-2]
    #   
    elif ls_y[-3] < ls_y[-2] and ls_y[-2] > y:
        x_max = ls_x[-2]
        flag += 1
        function_up[flag] = [x_min, x_max]

print(' ,    50  100:')
for key, val in function_up.items():
    print(f' {key}:     =  {val[0]}  {val[1]}')
```

```python
"""  ,     """

x = np.linspace(50, 100, 1000)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30

ls_x = []
ls_y = []
function_up = {}
flag = 0
x_max = 0
for x, y in zip(x, y):
    ls_x.append(x)
    ls_y.append(y)
    if len(ls_y) < 3:
        continue
    #   
    elif ls_y[-3] < ls_y[-2] and ls_y[-2] > y:
        x_max = ls_x[-2]
    #   
    elif ls_y[-3] > ls_y[-2] and ls_y[-2] < y:
        x_min = ls_x[-2]
        if not x_max:
            continue
        flag += 1
        function_up[flag] = [x_max, x_min]

print(' ,    50  100.     :')
for key, val in function_up.items():
    print(f' {key}:     =  {val[0]}  {val[1]}')
```

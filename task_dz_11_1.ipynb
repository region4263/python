```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplots()

x = np.linspace(-100, 100, 200)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30
plt.plot(x, y , label='функция')

plt.xlabel('ось X')
plt.ylabel('ось Y')
plt.title('вычисление функции')
plt.legend()

plt.show()
```

```python
""" Определить промежутки, на которых: f > 0, f < 0 """

x = np.linspace(-100, 100, 20000)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30

y_masked = np.ma.masked_where(y < 0, y)
plt.plot(x, y_masked, label='функция > 0')
y_masked = np.ma.masked_where(y > 0, y)
plt.plot(x, y_masked, label='функция < 0')

plt.legend()
plt.show()
```

```python
""" Вычислить вершину """

x = np.linspace(0, 100, 10000)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30
plt.plot(x, y, label='функция от 0 до 100')

max_y = max(y)
for x, y in zip(x, y):
    if y == max_y:
        print(f' На промежутке х[0: 100] вершина равна: x={x}, y={y}')
        plt.scatter(x, y, color='r', label='вершина')

plt.legend()
plt.show()
```

```python
""" Найти интервалы, на которых функция возрастает """

x = np.linspace(50, 100, 1000)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30

ls_x = []
ls_y = []
function_up = {}
flag = 0
for x, y in zip(x, y):
    ls_x.append(x)
    ls_y.append(y)
    if len(ls_y) < 3:
        continue
    # минимальная точка интервала
    elif ls_y[-3] > ls_y[-2] and ls_y[-2] < y:
        x_min = ls_x[-2]
    # максимальная точка интервала
    elif ls_y[-3] < ls_y[-2] and ls_y[-2] > y:
        x_max = ls_x[-2]
        flag += 1
        function_up[flag] = [x_min, x_max]

print('Рассмотрим промежуток, где Х от 50 до 100:')
for key, val in function_up.items():
    print(f'интервал №{key}: функция возрастает при Х = от {val[0]} до {val[1]}')
```

```python
""" Найти интервалы, на которых функция убывает """

x = np.linspace(50, 100, 1000)
y = -12 * pow(x, 4) * np.sin((np.cos(x))) - 18 * pow(x, 3) + 5 * pow(x, 2) + 10 * x - 30

ls_x = []
ls_y = []
function_up = {}
flag = 0
x_max = 0
for x, y in zip(x, y):
    ls_x.append(x)
    ls_y.append(y)
    if len(ls_y) < 3:
        continue
    # максимальная точка интервала
    elif ls_y[-3] < ls_y[-2] and ls_y[-2] > y:
        x_max = ls_x[-2]
    # минимальная точка интервала
    elif ls_y[-3] > ls_y[-2] and ls_y[-2] < y:
        x_min = ls_x[-2]
        if not x_max:
            continue
        flag += 1
        function_up[flag] = [x_max, x_min]

print('Рассмотрим промежуток, где Х от 50 до 100. Функция убывает на следующих интервалах:')
for key, val in function_up.items():
    print(f'интервал №{key}: функция убывает при Х = от {val[0]} до {val[1]}')
```
